### Import packages, set main window, create empty dictionary

```python
import tkinter as tk
from tkinter import messagebox

root = tk.Tk()
root.title("Contact list")
root.geometry("250x250")

contactbook = {}
```

### Set up the necessary functions and console:

```python

def addcontact():
    addwindow = tk.Toplevel(root)
    addwindow.title("Add contact details")

    tk.Label(addwindow, text="Name:").grid(row=0, column=0, padx=5, pady=5)
    newcontactname_entry = tk.Entry(addwindow)
    newcontactname_entry.grid(row=0, column=1, padx=5, pady=5)

    tk.Label(addwindow, text="Email:").grid(row=1, column=0, padx=5, pady=5)
    newcontactemail_entry = tk.Entry(addwindow)
    newcontactemail_entry.grid(row=1, column=1, padx=5, pady=5)

    tk.Label(addwindow, text="Number:").grid(row=2, column=0, padx=5, pady=5)
    newcontactnum_entry = tk.Entry(addwindow)
    newcontactnum_entry.grid(row=2, column=1, padx=5, pady=5)

    def add_to_contactbook():
        newcontactname = newcontactname_entry.get().title()
        newcontactemail = newcontactemail_entry.get().lower()
        newcontactnum = newcontactnum_entry.get()

        if newcontactname in contactbook:
            messagebox.showerror("Error", "Contact already exists!")
        else:
            contactbook[newcontactname] = {'email': newcontactemail, 'number': newcontactnum}
            messagebox.showinfo("Success", "Contact added successfully!")
            addwindow.destroy()

    tk.Button(addwindow, text="Add Contact", command=add_to_contactbook).grid(row=3, column=0, columnspan=2, padx=5, pady=5)

def searchcontact():
    searchwindow = tk.Toplevel(root)
    searchwindow.title("Search contact")
    tk.Label(searchwindow, text="Name:").grid(row=0, column=0, padx=5, pady=5)
    searchname_entry = tk.Entry(searchwindow)
    searchname_entry.grid(row=0, column=1, padx=5, pady=5)

    def searchcontactcheck():
        searchname = searchname_entry.get().title()
        if searchname in contactbook:
            contact_info = contactbook[searchname]
            messagebox.showinfo("Contact Info", f"Name: {searchname}\nEmail: {contact_info['email']}\nNumber: {contact_info['number']}")
            searchwindow.destroy()
        else:
            messagebox.showinfo("Not found", "The following contact is not found in the contact book.")
    
    tk.Button(searchwindow, text="Search", command=searchcontactcheck).grid(row=1, column=0, columnspan=2, padx=5, pady=5)

def updatecontact():
    updatewindow = tk.Toplevel(root)
    updatewindow.title("Update contact")
    tk.Label(updatewindow, text="Name of the contact to update:").grid(row=0, column=0, padx=5, pady=5)
    oldname_entry = tk.Entry(updatewindow)
    oldname_entry.grid(row=0, column=1, padx=5, pady=5)

    def updatename():
        oldname = oldname_entry.get().title()
        if oldname in contactbook:
            newname = newname_entry.get().title()
            contactbook[newname] = contactbook.pop(oldname)
            messagebox.showinfo("Success", "Contact name updated successfully.")
            updatewindow.destroy()
        else:
            messagebox.showerror("Error", "Contact not found.")


    def updateemail():
        oldname = oldname_entry.get().title()
        if oldname in contactbook:
            newemail = newemail_entry.get().lower()
            contactbook[oldname]['email'] = newemail
            messagebox.showinfo("Success", "Contact email updated successfully.")
            updatewindow.destroy()
        else:
            messagebox.showerror("Error", "Contact not found.")

    def updatenumber():
        oldname = oldname_entry.get().title()
        if oldname in contactbook:
            newnumber = newnumber_entry.get()
            contactbook[oldname]['number'] = newnumber
            messagebox.showinfo("Success", "Contact number updated successfully.")
            updatewindow.destroy()
        else:
            messagebox.showerror("Error", "Contact not found.")


    tk.Label(updatewindow, text="New Name:").grid(row=1, column=0, padx=5, pady=5)
    newname_entry = tk.Entry(updatewindow)
    newname_entry.grid(row=1, column=1, padx=5, pady=5)
    tk.Button(updatewindow, text="Update Name", command=updatename).grid(row=1, column=2, padx=5, pady=5)

    tk.Label(updatewindow, text="New Email:").grid(row=2, column=0, padx=5, pady=5)
    newemail_entry = tk.Entry(updatewindow)
    newemail_entry.grid(row=2, column=1, padx=5, pady=5)
    tk.Button(updatewindow, text="Update Email", command=updateemail).grid(row=2, column=2, padx=5, pady=5)

    tk.Label(updatewindow, text="New Number:").grid(row=3, column=0, padx=5, pady=5)
    newnumber_entry = tk.Entry(updatewindow)
    newnumber_entry.grid(row=3, column=1, padx=5, pady=5)
    tk.Button(updatewindow, text="Update Number", command=updatenumber).grid(row=3, column=2, padx=5, pady=5)

def deletecontact():
    deletewindow = tk.Toplevel(root)
    deletewindow.title("Delete contact")
    tk.Label(deletewindow, text="Name of the contact to delete:").grid(row=0, column=0, padx=5, pady=5)
    oldname_entry = tk.Entry(deletewindow)
    oldname_entry.grid(row=0, column=1, padx=5, pady=5)

    def delete():
        oldname = oldname_entry.get().title()
        if oldname in contactbook:
            del contactbook[oldname]
            messagebox.showinfo("Success", "Contact deleted successfully.")
            deletewindow.destroy()
        else:
            messagebox.showerror("Error", "Contact not found.")

    tk.Button(deletewindow, text="Delete Contact", command=delete).grid(row=1, column=0, columnspan=2, padx=5, pady=5)


def exportcontacts():
    textfile = "Contactbook.txt"
    with open(textfile, "w") as file:
        for name, info in contactbook.items():
            file.write(f"Name: {name}\n")
            file.write(f"Email: {info['email']}\n")
            file.write(f"Number: {info['number']}\n")
            file.write("\n")  # Add a newline to separate each contact

    messagebox.showinfo("Export Successful", f"Contacts exported to {textfile} successfully.")

add_button = tk.Button(root, text="Add Contact", command=addcontact)
add_button.grid(row=1, column=1, columnspan=2, padx=50, pady=5)

search_button = tk.Button(root, text="Search Contact", command=searchcontact)
search_button.grid(row=2, column=1, columnspan=2, padx=50, pady=5)

update_button = tk.Button(root, text="Update Contact", command=updatecontact)
update_button.grid(row=3, column=1, columnspan=2, padx=50, pady=5)

delete_button = tk.Button(root, text="Delete Contact", command=deletecontact)
delete_button.grid(row=4, column=1, columnspan=2, padx=50, pady=5)

export_button = tk.Button(root, text="Export All Contacts to Textfile", command=exportcontacts)
export_button.grid(row=5, column=1, columnspan=2, padx=50, pady=5)

quit_button = tk.Button(root, text="Quit", command=root.destroy)
quit_button.grid(row=6, column=1, columnspan=2, padx=50, pady=5)

root.mainloop()
```
